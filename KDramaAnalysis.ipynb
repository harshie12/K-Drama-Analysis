## **OUTLIER DETECTION AND TREATMENT**

```python
#Detecting and romoving outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]

#to get the dimensions of a DataFrame
df.shape
```

## **DATA PREPROCESSING**

```python
# Load or create your dataset (x: features, y: target variable)
x = df[['Drama_Name', 'Ratings', 'Actors', 'Viewship_Rate', 'Genre', 'Director','Writer', 'Year', 'Episodes', 'Network']]#2-D df, independent
x
```

```python
# Load or create your dataset (x: features, y: target variable)
y = df.iloc[ : , 0]#1-D series, dependent
y
```

```python
# Display the first few rows of the dataset(If there is no specific number given, the first 5 rows are shown by default)
x.head()
```

```python
# Checking the datatypes
x.dtypes
```

## **ENCODING(ORDINAL)**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
x[["Drama_Name", "Actors", "Genre", "Director", "Writer", "Network"]] = oe.fit_transform(x[["Drama_Name", "Actors", "Genre", "Director", "Writer", "Network"]])
#fit and transform are two different method
#fit = to perform calculations on the selected data
#transform = applies that calculation
```

```python
# Display the first few rows of the dataset(If there is no specific number given, the first 5 rows are shown by default)
x.head()
```

## **LOGISTIC REGRESSION**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42,)
```

```python
# Initialize the logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(xtrain, ytrain)

# Make predictions on the test data
ypred = model.predict(xtest)
```

```python
# Classification report
print(classification_report(ytest, ypred))
```

```python
# Confusion Matrix
print(confusion_matrix(ytest, ypred))
```

```python
# calculates the accuracy of the logistic regression model on the test dataset (Xtrain and ytrain)
model.score(xtrain, ytrain)
```

```python
# calculates the accuracy of the logistic regression model on the test dataset (Xtest and ytest)
model.score(xtest, ytest)
```

## **INSIGHTS**

```python
# Average rating of K-dramas by release season
season_rating = df.groupby('Year')['Ratings'].mean()
print(season_rating)
```

```python
# Plot the average rating of K-dramas by release season
sns.barplot(x=season_rating.index, y=season_rating.values)
plt.xlabel('Year')
plt.ylabel('Ratings')
plt.title('Average Rating of K-dramas by Release Year')
plt.show()
```

```python
# Number of K-dramas produced each year
yearly_drama_count = df['Year'].value_counts().sort_index()
print(yearly_drama_count)
```

```python
# Plot the number of K-dramas produced each year
plt.plot(yearly_drama_count.index, yearly_drama_count.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of K-dramas')
plt.title('Number of K-dramas Produced Each Year')
plt.show()
```
